```python
param_list = ['Country',
              'Graduates in science and engineering, %_2022',
              'Gross expenditure on R&D, % GDP_2022',
              'High-tech exports, % total trade_2022',
              'High-tech imports, % total trade_2022',
              'High-tech manufacturing, %_2022',
              'Information and Communications Technology access_2022',
              'Information and Communications Technology services exports, % total trade_2022',
              'Information and Communications Technology services imports, % total trade_2022',
              'Information and Communications Technology use_2022',
              'Knowledge absorption_2022',
              'PISA scales in reading, maths and science_2022',
              'Pupil-teacher ratio, secondary_2022',
              'Knowledge workers_2022',
              'Knowledge and technology outputs index_2022',
              'Income',
              'High income']

income_mapping = {
    "low income": 0,
    "middle income": 1,
    "high income": 2
}

short_names = {
    'Graduates in science and engineering, %_2022': 'STEM Graduates',
    'Gross expenditure on R&D, % GDP_2022': 'R&D Exp (% GDP)',
    'High-tech exports, % total trade_2022': 'Hi-Tech Exp',
    'High-tech imports, % total trade_2022': 'Hi-Tech Imp',
    'High-tech manufacturing, %_2022': 'Hi-Tech Mfg',
    'Information and Communications Technology access_2022': 'ICT Access',
    'Information and Communications Technology services exports, % total trade_2022': 'ICT Serv Exp',
    'Information and Communications Technology services imports, % total trade_2022': 'ICT Serv Imp',
    'Information and Communications Technology use_2022': 'ICT Use',
    'Knowledge absorption_2022': 'Knowl. Absorption',
    'PISA scales in reading, maths and science_2022': 'PISA Score',
    'Pupil-teacher ratio, secondary_2022': 'Pupil-Teacher Ratio',
    'Knowledge workers_2022': 'Knowl. Workers',
    'Knowledge and technology outputs index_2022': 'Knowl-Tech Index',
    'Income_rank': 'Income Rank',
    'High income': 'High Income (bin)'
}

df_final = (
    pl.read_excel('/content/drive/MyDrive/Colab Notebooks/WIPO-GII_2022.xlsx')
    .select(param_list)
    .with_columns([
        pl.col("Income")
        .str.to_lowercase()
        .str.strip_chars()
        .replace_strict(income_mapping, default=None)
        .cast(pl.Int8)
        .alias("Income_rank"),

        pl.col("High income").cast(pl.Int8)
    ])
    .drop_nulls()
)


numeric_df = df_final.select(cs.numeric())
column_names = numeric_df.columns

corr_matrix = numeric_df.with_columns(
    pl.all().rank()
).corr()

corr_long = (
    corr_matrix
    .with_columns(var1 = pl.Series(column_names))
    .unpivot(index="var1", variable_name="var2", value_name="correlation")
    .with_columns([
        pl.col("var1").replace_strict(short_names, default=pl.col("var1")).alias("var1_short"),
        pl.col("var2").replace_strict(short_names, default=pl.col("var2")).alias("var2_short")
    ])
)

base = alt.Chart(corr_long).encode(
    x=alt.X('var1_short:N', title=None, axis=alt.Axis(labelAngle=-45, labelLimit=150)),
    y=alt.Y('var2_short:N', title=None, axis=alt.Axis(labelLimit=150)),
    tooltip=[
        alt.Tooltip('var1:N', title='Полное имя 1'),
        alt.Tooltip('var2:N', title='Полное имя 2'),
        alt.Tooltip('correlation:Q', format='.2f')
    ]
)

heatmap = base.mark_rect().encode(
    color=alt.Color('correlation:Q',
                    scale=alt.Scale(scheme='redblue', domain=[-1, 1]),
                    title="r-Спирмена")
)

text = base.mark_text(fontSize=9).encode(
    text=alt.Text('correlation:Q', format='.2f'),
    color=alt.condition(
        "abs(datum.correlation) > 0.5",
        alt.value('white'),
        alt.value('black')
    )
)

(heatmap + text).properties(
    width=650, height=600,
    title="Матрица корреляций"
).display()
```

3. Выберите подходящий алгоритм кластеризации и обоснуйте свой выбор. Определите число кластеров, объясните своё решение. Опишите получившиеся кластеры с помощью описательной статистки. Нужно не просто рассчитать статистики в таблицах, но детально описать полученные результаты. Дайте названия кластерам и обоснуйте их.

```python
X_clustering = df_pca_final.select(["PC1 - Технологическое превосходство", "PC2 - Хай-тек торговля", "PC3 - Потенциал развития"]).to_numpy()

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_clustering)
    inertia.append(kmeans.inertia_)

# 3. Создаем график для визуализации
elbow_data = pl.DataFrame({
    "k": list(k_range),
    "inertia": inertia
})

elbow_chart = alt.Chart(elbow_data).mark_line(point=True).encode(
    x=alt.X('k:O', title='Количество кластеров (k)'),
    y=alt.Y('inertia:Q', title='Инерция (Inertia)'),
    tooltip=['k', 'inertia']
).properties(
    width=600,
    height=400,
    title="Метод локтя для определения оптимального числа кластеров"
)

elbow_chart.display()
```

```python
# 1. Готовим данные (те же 3 компоненты)
# Используем названия, которые у тебя сейчас в памяти (длинные)
features = ["PC1 - Технологическое превосходство",
            "PC2 - Хай-тек торговля",
            "PC3 - Потенциал развития"]
X_hierarchy = df_pca_final.select(features).to_numpy()
country_names = df_pca_final.get_column("Country").to_list()

# 2. Вычисляем матрицу связей (linkage)
# Метод 'ward' минимизирует дисперсию внутри кластеров (хорошо сочетается с K-Means)
Z = linkage(X_hierarchy, method='ward')

# 3. Отрисовка
plt.figure(figsize=(20, 10))
plt.title("Иерархическая кластеризация стран (Дендрограмма)", fontsize=16)
plt.xlabel("Страны", fontsize=12)
plt.ylabel("Расстояние (Сходство)", fontsize=12)

# Рисуем дерево
dendrogram(
    Z,
    labels=country_names,
    leaf_rotation=90.,  # Поворачиваем названия стран вертикально
    leaf_font_size=8.,  # Размер шрифта для названий
    color_threshold=7   # Порог цвета (поможет визуально выделить группы)
)

plt.axhline(y=7, color='r', linestyle='--') # Линия отсечки для кластеров
plt.tight_layout()
plt.show()
```

## **Анализ данных:**


**Определение оптимального числа кластеров**

Для выбора количества групп был использован метод «локтя»

Анализ графика: На графике зависимости инерции от количества кластеров наблюдается резкое снижение показателя при переходе от 1 к 2 и от 2 к 3 группам. После точки k=4 угол наклона кривой заметно сглаживается, что свидетельствует о затухании маржинального выигрыша от дальнейшего дробления выборки.

Решение: Оптимальным признано разделение на 4 кластера. Это позволяет сохранить баланс между детальностью описания и статистической значимостью каждой группы.


**Выбор алгоритма и обоснование**

Для сегментации стран был выбран алгоритм K-Means (метод k-средних).

**Обоснование:** Данный алгоритм является классическим инструментом для работы с метрическими данными после снижения размерности через PCA. Он эффективно минимизирует суммарное квадратичное отклонение точек внутри кластеров, что позволяет получить компактные и интерпретируемые группы стран в пространстве главных компонент.


```python
# 1. Обучаем модель K-Means на 4 кластера
kmeans_final = KMeans(n_clusters=4, random_state=42, n_init=10)
# Используем те же данные X_clustering, что готовили для графика
clusters = kmeans_final.fit_predict(X_clustering)

# 2. Добавляем результат в нашу таблицу
df_pca_final = df_pca_final.with_columns(
    Cluster = pl.Series(clusters).cast(pl.Utf8) # Делаем текст, чтобы графики не путали с числами
)

# 3. Смотрим, кто попал в кластер вместе с Россией
russia_cluster = df_pca_final.filter(pl.col("Country") == "Russian Federation").get_column("Cluster")[0]
neighbors = df_pca_final.filter(pl.col("Cluster") == russia_cluster).get_column("Country").to_list()

print(f"Россия попала в КЛАСТЕР №{russia_cluster}")
print(f"Её соседи по группе (первые 10): {neighbors[:10]}")
```

    Анализ: В этот кластер не попала ни одна страна с высоким уровнем дохода. Он состоит исключительно из стран с низким (13) и средним (30) доходом. Это математически доказывает, что низкие показатели технологического превосходства и цифровизации являются прямым следствием (и одновременно причиной) экономического отставания.

5. Почему получилась именно такая кластерная структура? Полученные результаты соответствуют вашем изначальным предположениям или оказались неожиданными?

## Ответ на вопрос:

**Почему получилась именно такая структура?**


Кластерная структура сформировалась под влиянием двух мощных факторов:

* Доминирование PC1 (54% вариации): PC1 в основном собирает в себе показатели уровня образования (PISA), расходов на R&D и доли knowledge workers / STEM-кадров. Из-за такого доминирования она сразу разделила страны на две большие группы:
- те, у кого все эти показатели высокие (кластеры 1 и 2),
- и все остальные (кластеры 0 и 3).
Поэтому граница между «сильными» и «слабыми» кластерами получилась очень чёткой

* Эффект «Специализации» (PC2 и PC3):
    * PC2 (Торговля) - это в основном различие по экспорту высокотехнологичной продукции (high-tech exports % от manufactured exports). Страны кластера 1 активно продают чипы, электронику, софт и т.д. на внешние рынки.
    * PC3 (Потенциал) выделила группу стран, которые пока не стали лидерами по доходу, но вкладывают в STEM и кадры непропорционально много для своего уровня развития (Кластер 3).

**Что соответствовало изначальным предположениям, а что нет**

Соответствие:

  * Корреляция с доходом: Было вполне предсказуемо, что страны с высоким домом (High Income) сконцентрируются в «сильных» кластерах (1 и 2), а страны с низким - в «слабом» (0). Математика лишь подтвердила экономическую аксиому: инновации требуют больших денег.

  * Лидерство США и Великобритании: Их попадание в топ-группы по PC1 было очевидным, так как эти страны исторически лидируют в рейтингах университетов и затратах на науку.


Оказалось неожиданным:

  * Существование Кластера №2: Неожиданным оказалось выделение 13 стран, у которых высокий показатель «Превосходства» (PC1 ~3.9), но при этом провал по «Торговле» (PC2 ~ -1.8). Это говорит о том, что высокий интеллект нации не всегда автоматически конвертируется в экспорт высокотехнологичных товаров.

  * Роль Ирана (PC3 = 2.19): Тот факт, что страны вроде Ирана показывают огромный отрыв по «Потенциалу развития», стал сюрпризом. (во многом именно это и определило название этой компоненты). Это ломает стереотип о том, что развивающиеся страны слабы во всём. В кадровом и инженерном плане (STEM) они могут конкурировать с лидерами.

  * Позиция России: Отрицательный PC3 (-0.63) при положительном PC1 (0.89) - это важный инсайт. Это указывает на структурный перекос: при наличии сильной научной базы (PC1) страна испытывает сложности с воспроизводством новых кадров и экспортом современных IT-сервисов (PC3).


**Итоговый вывод:**

Данные через PCA показали, что нет простой линейной схемы «богатство -> автоматически инновации». Есть разные траектории развития.
Самое ценное, на мой взгляд - кластер 3. Он говорит, что инвестиции в STEM-образование и подготовку инженеров/учёных можно (и нужно) начинать сильно раньше, чем страна становится богатой. Если этого не делать, то даже при росте экономики потом может не хватить квалифицированных людей, чтобы этот рост превратить в реальные технологии и экспорт.
