```python
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import scipy.stats as stats
import re
import statsmodels.formula.api as smf
```

```python
def clean_colname(name):
    """
    Превращает '2020 Exports of goods and services (% of GDP)'
    в 'exports_of_goods_and_services_percent_of_gdp'
    """
    # 1. Убираем год в начале (4 цифры + пробел)
    name = re.sub(r'^\d{4}\s*', '', str(name))
    
    # 2. Убираем спецсимволы: % $ ( ) ,
    name = re.sub(r'[%$(),]', '', name)
    
    # 3. Заменяем пробелы и множественные подчёркивания на одно _
    name = re.sub(r'\s+', '_', name)
    name = re.sub(r'_+', '_', name)
    
    # 4. В нижний регистр и убираем подчёркивания по краям
    return name.lower().strip('_')



def analyze_column(df, column_name):
    
    if column_name not in df.columns:
        print(f"⚠️ Колонка '{column_name}' не найдена!")
        return
    
    print(f"\n📊 Анализ колонки: {column_name}")
    print("-" * 60)
    
    # Тип данных
    print(f"Тип данных: {df[column_name].dtype}")
    
    # Общее количество наблюдений
    total = len(df)
    print(f"Всего наблюдений: {total}")
    
    # Пропуски
    missing = df[column_name].isna().sum()
    print(f"Количество пропусков (NaN): {missing}")
    
    # Нули (если числовая)
    if pd.api.types.is_numeric_dtype(df[column_name]):
        zeros = (df[column_name] == 0).sum()
        print(f"Количество нулевых значений: {zeros}")
    
    # Уникальные значения
    unique_vals = df[column_name].nunique()
    print(f"Количество уникальных значений: {unique_vals}")
    
    # Примеры значений
    print(f"Примеры значений: {df[column_name].dropna().head(10).tolist()}")
    
    # Статистика для числовых переменных
    if pd.api.types.is_numeric_dtype(df[column_name]):
        print("\n📈 Описательная статистика:")
        print(df[column_name].describe())
```

```python
columns = ['2020 Communications, computer, etc. (% of service exports, BoP)',
           '2020 Exports of goods and services (% of GDP)',
           '2020 Merchandise exports to high-income economies (% of total merchandise exports)', 
           '2020 Fuel exports (% of merchandise exports)', 
           '2020 Food exports (% of merchandise exports)']


for i, col in enumerate(columns):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    df = df[df[col].notna()]

    sns.histplot(df[col].dropna(),bins=30, stat="density", kde=True, ax=axes[0])
    
    mu, std = stats.norm.fit(df[col].dropna())
    xmin, xmax = axes[0].get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, std)
    axes[0].plot(x, p, 'k', linewidth=2)
    
    stats.probplot(df[col], dist="norm", plot=axes[1])
    axes[1].set_title('Q-Q Plot (Сравнение с нормальным распределением)')
```

```python
var = '2023 Communications, computer, etc. (% of service exports, BoP)'

group_oecd = df[df['OECD members'] == 1][var].dropna()
group_non = df[df['OECD members'] == 0][var].dropna()
print(f"Группа OECD: n={len(group_oecd)}, среднее={group_oecd.mean():.2f}, std={group_oecd.std():.2f}")
print(stats.shapiro(group_oecd))
print(f"Группа non-OECD: n={len(group_non)}, среднее={group_non.mean():.2f}, std={group_non.std():.2f}")
print(stats.shapiro(group_non))

print(stats.levene(group_oecd, group_non))

stats.ttest_ind(group_oecd, group_non, equal_var=True)

plt.figure(figsize=(6,4))
sns.boxplot(x='OECD members', y=var, data=df)
sns.swarmplot(x='OECD members', y=var, data=df, color=".25")
plt.title('X: Communications, computer, etc. по группам OECD')
plt.xlabel('OECD members (1=Yes, 0=No)')
plt.ylabel('% of service exports')
plt.show()
```

Результат t-теста:
Страны - члены ОЭСР имеют статистически значимо более высокую долю коммуникационных, компьютерных и прочих услуг в экспорте услуг по сравнению со странами, не входящими в ОЭСР (p < 0.05). 


```python
var_y = '2023 Transport services (% of service exports, BoP)'

group_oecd_y = df[df['OECD members'] == 1][var_y].dropna()
group_non_y = df[df['OECD members'] == 0][var_y].dropna()

print(stats.shapiro(group_oecd_y))
print(stats.shapiro(group_non_y))
print(stats.levene(group_oecd_y, group_non_y))

stats.mannwhitneyu(group_oecd_y, group_non_y, alternative='two-sided')

plt.figure(figsize=(6,4))
sns.boxplot(x='OECD members', y=var_y, data=df)
sns.swarmplot(x='OECD members', y=var_y, data=df, color=".25")
plt.title('Y: Transport services по группам OECD')
plt.xlabel('OECD members (1=Yes, 0=No)')
plt.ylabel('% of service exports')
plt.show()
```

_____________________________________________________________________________________
Тест Mann-Whitney (U-тест) не показал статистически значимых различий между странами ОЭСР и остальными странами по доле транспортных услуг в экспорте услуг в 2023 году.

Статистически значимых различий между странами ОЭСР и остальными странами по доле транспортных услуг в экспорте услуг не выявлено (p = 0.795).
Это означает, что в отличие от высокотехнологичных и коммуникационных услуг, транспортные услуги в структуре экспорта услуг в равной мере характерны как для развитых, так и для развивающихся стран.

```python
var = '2023 Communications, computer, etc. (% of service exports, BoP)'

for name, group in df.groupby('Income'):
    print(name, stats.shapiro(group[var].dropna()))
```

____________________________________________________________________
Так как тест Шапиро не показал нормальность на данных группированных по среднему доходу тест ANOVA не подойдет для этих данных. Поэтому мы используем Kruskal-Wallis H-test.

```python
groups = [df[df['Income'] == inc]['2023 Communications, computer, etc. (% of service exports, BoP)'].dropna().values
          for inc in ['low income', 'middle income', 'high income']]

stat, p = stats.kruskal(*groups)
print(f"Kruskal-Wallis H-statistic: {stat:.3f}, p-value: {p:.3f}")

plt.figure(figsize=(8,5))
sns.boxplot(x='Income', y=var, data=df, order=['low income','middle income','high income'])
sns.swarmplot(x='Income', y=var, data=df, color=".25", order=['low income','middle income','high income'])
plt.title('X: Communications, computer, etc. по группам дохода')
plt.xlabel('Income group')
plt.ylabel('% of service exports')
plt.show()
```

________________________________________________
Тест Крускала-Уоллиса показал, что различия в доле экспорта услуг связи и IT между странами с разным уровнем дохода не являются статистически значимыми (H = 4.02, p = 0.134). Это говорит о том, что доходная группа страны не сильно влияет на долю этих услуг в общем экспорте.

```python
for name, group in df.groupby('Income'):
    print(name, stats.shapiro(group[var_y].dropna()))
    
groups_y = [df[df['Income'] == inc]['2023 Transport services (% of service exports, BoP)'].dropna().values
          for inc in ['low income', 'middle income', 'high income']]
stat, p = stats.kruskal(*groups_y)
print(f"Kruskal-Wallis H-statistic: {stat:.3f}, p-value: {p:.3f}")

plt.figure(figsize=(8,5))
sns.boxplot(x='Income', y=var, data=df, order=['low income','middle income','high income'])
sns.swarmplot(x='Income', y=var, data=df, color=".25", order=['low income','middle income','high income'])
plt.title('X: Communications, computer, etc. по группам дохода')
plt.xlabel('Income group')
plt.ylabel('% of service exports')
plt.show()
```

___________________________________________________________________
Доля транспортных услуг в общем экспорте не отличается существенно между странами с разным уровнем дохода в 2023 году (H = 0.815, p = 0.665). Независимо от того, страна с высоким, средним или низким доходом, структура экспорта транспортных услуг примерно одинакова.

```python
var_x = '2023 Communications, computer, etc. (% of service exports, BoP)'
var_y = '2023 Transport services (% of service exports, BoP)'

# Список регионов с реальными данными
regions = df['Region2'].dropna().unique()

# Группы для X
groups_x = [df[df['Region2'] == r][var_x].dropna().values for r in regions]

# Kruskal-Wallis тест для X
stat_x, p_x = stats.kruskal(*groups_x)
print(f"Kruskal-Wallis X: H={stat_x:.3f}, p={p_x:.3f}")

# Группы для Y
groups_y = [df[df['Region2'] == r][var_y].dropna().values for r in regions]

# Kruskal-Wallis тест для Y
stat_y, p_y = stats.kruskal(*groups_y)
print(f"Kruskal-Wallis Y: H={stat_y:.3f}, p={p_y:.3f}")

# Визуализация X
plt.figure(figsize=(12,5))
sns.boxplot(x='Region2', y=var_x, data=df)
plt.xticks(rotation=45)
plt.title('Распределение X по регионам')
plt.show()

# Визуализация Y
plt.figure(figsize=(12,5))
sns.boxplot(x='Region2', y=var_y, data=df)
plt.xticks(rotation=45)
plt.title('Распределение Y по регионам')
plt.show()
```

____________________________________________________________
Для сравнения распределения двух видов услуг по регионам (Region2) был использован непараметрический критерий Краскела-Уоллиса.

**1. Communications, computer, etc. (% of service exports, BoP)**

Результат теста Краскела-Уоллиса:
H = 10.401, p = 0.034 статистически значимые различия между регионами (p < 0.05).
Интерпретация боксплота:
Самый высокий медианный уровень наблюдается в Europe.
Высокие показатели также в регионах Americas и Asia-Pacific.
Самые низкие значения - в Sub-Saharan Africa и Middle East and North Africa.
Европа заметно лидирует, в то время как Африка и Ближний Восток значительно отстают.

Вывод:
Доля современных цифровых, коммуникационных и компьютерных услуг в экспорте услуг сильно зависит от региона. Развитые регионы (прежде всего Европа) значительно сильнее развиты в этом виде услуг.

**2. Transport services (% of service exports, BoP)**

Результат теста Краскела-Уоллиса:
H = 3.166, p = 0.530 статистически значимых различий между регионами нет (p > 0.05).
Интерпретация боксплота:
Медианные значения во всех регионах довольно близки.
Нет явного регионального лидера.
Во всех регионах наблюдается большой разброс значений (длинные «усы» и выбросы).

Вывод:
Доля транспортных услуг в экспорте услуг распределена относительно равномерно по регионам мира. Этот вид услуг в меньшей степени зависит от уровня экономического развития региона и в большей степени - от географического положения, наличия портов, логистических коридоров и транзитного потенциала.

```python
var_x = 'Communications, computer, etc. (% of service exports, BoP)'
var_y = 'Transport services (% of service exports, BoP)'

# Выровнять по индексам и убрать NaN одновременно
x_22_23 = df[[f'2022 {var_x}', f'2023 {var_x}']].dropna()
y_22_23 = df[[f'2022 {var_y}', f'2023 {var_y}']].dropna()

# Разности
diff_x = x_22_23[f'2023 {var_x}'] - x_22_23[f'2022 {var_x}']
diff_y = y_22_23[f'2023 {var_y}'] - y_22_23[f'2022 {var_y}']

# Shapiro-Wilk
w_x, p_x = stats.shapiro(diff_x)
w_y, p_y = stats.shapiro(diff_y)
print(f"Shapiro-Wilk для X (разности): W={w_x:.3f}, p={p_x:.3f}")
print(f"Shapiro-Wilk для Y (разности): W={w_y:.3f}, p={p_y:.3f}")

# Wilcoxon
stat, p = stats.wilcoxon(x_22_23[f'2023 {var_x}'], x_22_23[f'2022 {var_x}'])
print(f"Wilcoxon signed-rank X: stat={stat:.3f}, p={p:.3f}")
stat, p = stats.wilcoxon(y_22_23[f'2023 {var_y}'], y_22_23[f'2022 {var_y}'])
print(f"Wilcoxon signed-rank Y: stat={stat:.3f}, p={p:.3f}")

plt.figure(figsize=(10,5))
sns.boxplot(data=df[['2022 Communications, computer, etc. (% of service exports, BoP)',
                     '2023 Communications, computer, etc. (% of service exports, BoP)']])
plt.title("X: Communications/Computer Services 2022 vs 2023")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(data=df[['2022 Transport services (% of service exports, BoP)',
                     '2023 Transport services (% of service exports, BoP)']])
plt.title("Y: Transport Services 2022 vs 2023")
plt.show()
```

__________________________________________________________________
Тест Шапиро-Уилка на нормальность разностей, показал отрицательный результат, поэтому мы используем непараметрический критерий Вилкоксона в обоих случаях.

**1) Сектор X: Коммуникационные и компьютерные услуги**

Этот сектор показывает высокую долю в экспорте (медиана около 40%). Статистика: $p=0.045$.

Вывод: Существует статистически значимое различие между 2022 и 2023 годами (так как $p < 0.05$).

Визуальный анализ: Если присмотреться к боксплотам, медиана в 2023 году (оранжевый график) находится чуть выше, чем в 2022 (синий). Разница небольшая, но на большой выборке она подтверждается математически. Сектор демонстрирует слабую тенденцию к росту или перераспределению долей.

**2) Сектор Y: Транспортные услуги**

Здесь доля в экспорте заметно ниже, но изменения более выражены.
Статистика: $p=0.000$.

Вывод: Различия крайне значимы.

Визуальный анализ: На графике видно заметное снижение медианы и межквартильного размаха в 2023 году по сравнению с 2022. Особенности: В 2022 году наблюдалось много выбросов (отдельные страны с аномально высокой долей транспортных услуг - до 80%). В 2023 году количество и амплитуда этих выбросов сократились. Это может указывать на общее снижение роли транспортного экспорта в структуре услуг или на стабилизацию рынка после шоков.
