```python
df = pd.read_excel('Worldbank_panel.xlsx')
print(df.info())
print(df.describe())
print(df.isnull().sum())
df.head(25)
```

```python
countries = sorted(df['id'].unique())

keys = ['v30', 'v6', 'v9', 'v13', 'v19', 'v21', 'v22', 'v24']

df[keys] = df.groupby('id')[keys].transform(lambda x: x.interpolate(method='linear', limit_direction='both'))

for col in keys:
    df[col] = df[col].fillna(df.groupby('year')[col].transform('mean'))

labels = {
    'v30': 'ВВП на душу населения (2015$)',
    'v6': 'Фиксированный ШПД (на 100 чел.)',
    'v9': 'Высокотехнологичный экспорт (%)',
    'v13': 'Экспорт ИКТ-услуг (%)',
    'v19': 'Мобильная связь (на 100 чел.)',
    'v21': 'Патенты резидентов',
    'v22': 'Научные статьи',
    'v24': 'Защищённые серверы (на 1 млн чел.)'
}
def plot_country(country, selected_keys, normalize=True):
    
    df_c = df[df['id'] == country].sort_values('year')
    
    plt.figure(figsize=(10,5))
    
    for key in selected_keys:
        y = df_c[key]
        
        if normalize:
            if y.max() != y.min():
                y = (y - y.min()) / (y.max() - y.min())
        
        plt.plot(df_c['year'], y, marker='o', label=labels[key])
    
    plt.title(f'Динамика для {country}')
    plt.xticks(df_c['year'])
    plt.xlabel('Год')
    plt.ylabel('Нормализованные значения' if normalize else 'Исходные значения')
    
    plt.legend()
    plt.grid(True)
    plt.show()
    
interact(
    plot_country,
    country=countries,
    selected_keys=SelectMultiple(
        options=keys,
        value=['v30', 'v6', 'v9', 'v13', 'v19', 'v21', 'v22', 'v24'],
        description='Показатели'
    )
)
```

```python
# 1. Рассчитаем средние значения показателей для каждой страны за все года
pivot_means = df.groupby('id')[keys].mean()

# 2. Функция для расчета роста ВВП
def calculate_growth(group):
    # Берем значения за конкретные годы
    v30_2015 = group.loc[group['year'] == 2015, 'v30'].values
    v30_2018 = group.loc[group['year'] == 2018, 'v30'].values
    # Проверяем, что данные есть и не делим на ноль
    if len(v30_2015) > 0 and len(v30_2018) > 0 and v30_2015[0] != 0:
        return ((v30_2018[0] - v30_2015[0]) / v30_2015[0]) * 100
    return None

# Группируем и применяем расчет роста
growth = df.groupby('id').apply(calculate_growth, include_groups=False)

# 3. Собираем итоговую таблицу
summary_table = pivot_means.copy()
summary_table['Рост ВВП 2015-2018 (%)'] = growth

# 4. Сортируем по убыванию роста ВВП
summary_table = summary_table.sort_values(by='Рост ВВП 2015-2018 (%)', ascending=False)

# 1. Готовим словарь для переименования
display_labels = labels.copy()
display_labels['Рост ВВП 2015-2018 (%)'] = 'Темп роста экономики (%)'

# 2. Применяем переименование к нашей сводной таблице
# summary_table уже содержит средние значения и расчет роста
final_table = summary_table.rename(columns=display_labels)

# 3. Выводим результат
print("ИТОГОВАЯ ТАБЛИЦА СРАВНЕНИЯ СТРАН (2015-2018)")
display(final_table.head(10))
```

```python
country_stats = df.groupby('id')[keys].mean()
descriptive_table = country_stats.describe().rename(columns=labels)
display(descriptive_table)
```

```python
# Логарифмируем показатели для более адекватного сравнения
for col in keys:
    df[f'l_{col}'] = np.log1p(df[col])
    
df_stats = df[['id', 'year', 'l_v30', 'l_v6', 'l_v9', 'l_v13', 'l_v19', 'l_v21', 'l_v22', 'l_v24']]

year_dummies = pd.get_dummies(df_stats['year'], prefix='year', drop_first=True).astype(int)

df_stats = pd.concat([df_stats, year_dummies], axis=1)

df_stats.head()
```

```python
df_panel = df_stats.set_index(['id', 'year'])

X = sm.add_constant(df_panel.drop(columns=['l_v30']))

mod = PooledOLS(df_panel.l_v30, X)
pooled_res = mod.fit()
print(pooled_res)
```

```python
mod_re = RandomEffects(df_panel.l_v30, X)
re_res = mod_re.fit()
print(re_res)
```

```python
pd.DataFrame({
    "Coef": re_res.params,
    "P-value": re_res.pvalues,
    "exp(Coef)": np.exp(re_res.params)
})
```

```python
import numpy.linalg as la
from scipy import stats
def hausman(fe, re):
    b = fe.params
    B = re.params
    v_b = fe.cov
    v_B = re.cov
    df = b[np.abs(b) < 1e8].size
    chi2 = np.dot((b - B).T, la.inv(v_b - v_B).dot(b - B)) 
    pval = stats.chi2.sf(chi2, df)
    return chi2, df, pval

hausman_results = hausman(fe_res, re_res) 
print('p-value: ' + str(hausman_results[2]))
```
